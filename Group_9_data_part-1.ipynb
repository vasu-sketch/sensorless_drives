```python
# Almost all the columns are having outliers. Replacing outliers using the IQR method
# replacing them with min,max values
```

```python

#replacing outliers 
for cl in df.columns:
    q_75,q_25 = np.percentile(df.loc[:,cl],[75,25])
    intr_qr = q_75-q_25
 
    max = q_75+(1.5*intr_qr)
    min = q_25-(1.5*intr_qr)
 
    df.loc[df[cl] < min,cl] = min
    df.loc[df[cl] > max,cl] = max
```
